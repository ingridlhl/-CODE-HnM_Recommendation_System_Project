```python
b=df1[df1.day==0].groupby(df1.article_id).price.sum().sort_values(ascending=False)[0:10].index.values
df2=df1[df1.day==0][df1.article_id.isin(b)]
for i in range(1,734):
    b=df1[df1.day==i].groupby(df1.article_id).price.sum().sort_values(ascending=False)[0:10].index.values
    df_temp=df1[df1.day==i][df1.article_id.isin(b)]
    df2=pd.concat([df2,df_temp])
df2
```

```python
df3=df2.groupby(['day','article_id']).price.sum().reset_index()
```

```python
df3
```

```python
df3.article_id=df3.article_id.astype('str')
```

```python
df4=pd.get_dummies(df3,columns=['article_id'])
```

```python
df5=df4.groupby('day').sum()
```

```python
df5
```

```python
df6=df5.drop(columns='price')
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
train_size = int(len(df6) * 0.75) 
test_size = len(df6) - train_size
train = df6.iloc[0:train_size, :]
```

```python
train
```

```python
X_train = []
y_train = []
for i in range(60, 550):
    X_train.append(train.iloc[i-60:i, :]) 
    y_train.append(train.iloc[i, :])
X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
X_train.shape
```

```python
regressor = Sequential()

regressor.add(LSTM(units = 50, return_sequences = True, input_shape = (X_train.shape[1],X_train.shape[2])))
regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(LSTM(units = 50))
regressor.add(Dropout(0.2))
regressor.add(Dense(units = y_train.shape[1],activation='softmax'))
#regressor.add(Activation('sigmoid'))
regressor.compile(optimizer = 'adam', loss = 'categorical_crossentropy')

regressor.fit(X_train, y_train, epochs = 10, batch_size = 32)
```
